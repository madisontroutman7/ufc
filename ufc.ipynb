```python
import numpy as np
import pandas as pd
import json
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("data.csv")
```

```python
df['B_Age'] = pd.to_numeric(df['B_Age'], errors='coerce')

df = df.dropna(subset=['B_Age'])

sorted_df = df.sort_values(by='B_Age')

youngest_fighter = sorted_df.iloc[0]
oldest_fighter = sorted_df.iloc[-1]
middle_index = len(sorted_df) // 2
middle_age_fighter = sorted_df.iloc[middle_index]

print(f"Youngest Fighter: {youngest_fighter['B_Name']}, Age: {youngest_fighter['B_Age']}")
print(f"Oldest Fighter: {oldest_fighter['B_Name']}, Age: {oldest_fighter['B_Age']}")
print(f"Middle-Aged Fighter: {middle_age_fighter['B_Name']}, Age: {middle_age_fighter['B_Age']}")
```

```python
lower_age = 21
upper_age = 21

similar_age_fights = df[(df["B_Age"].between(lower_age, upper_age)) | (df["R_Age"].between(lower_age, upper_age))]

fighter_counts = pd.concat([similar_age_fights["B_Name"], similar_age_fights["R_Name"]]).value_counts()

top_10_fighters = fighter_counts.head(10).index.tolist()

blue_corner_weights = df[df["B_Name"].isin(top_10_fighters)][["B_Name", "B_Weight"]].rename(columns={"B_Name": "Fighter", "B_Weight": "Weight"})
red_corner_weights = df[df["R_Name"].isin(top_10_fighters)][["R_Name", "R_Weight"]].rename(columns={"R_Name": "Fighter", "R_Weight": "Weight"})

weights = pd.concat([blue_corner_weights, red_corner_weights])

weights = weights.groupby("Fighter")["Weight"].mean().sort_values(ascending=False)
weights.rename_axis("Top 10 fighters that are similar weight classes to Song Yadong (Age 21)", inplace=True)

print(weights)
```
